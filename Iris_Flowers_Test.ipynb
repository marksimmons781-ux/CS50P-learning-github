```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


iris = datasets.load_iris()

#print(iris)

features = iris.data

labels = iris.target

#print(features)
#print(labels)

features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size=.5)
print(len(features_train))

my_classifier = KNeighborsClassifier()
my_classifier.fit(features_train, labels_train)

prediction = my_classifier.predict (features_test)
print(prediction)
print(accuracy_score(labels_test, prediction))

iris1 = [[1.1, 3.1, 4.9, 1.0]] # Added a fourth feature (e.g., petal width)
iris_predictions = my_classifier.predict(iris1)
print(iris_predictions)

if iris_predictions == 0:
    print("Setosa")
if iris_predictions == 1:
    print("Versicolor")
if iris_predictions == 2:
    print("Virginica")
```
