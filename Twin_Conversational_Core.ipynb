```python
# --- PROJECT: TWIN - THE BOSTONIAN (REFINED US MALE) ---
# Location: /02_PYTHON_RPI_CODE/RPI_MAIN_PROGRAM/twin_core_v2.py

!pip install librosa soundfile gTTS
import random
import librosa
import soundfile as sf
from gtts import gTTS
import IPython.display as ipd
import time

# 1. REFINED VOICE ENGINE (The "Southie" Calibration)
def speak_as_twin(text):
    """Generates US Male speech and shifts pitch for a deep, gritty Boston tone."""
    # STRICT US English (Top Level Domain .com)
    tts = gTTS(text=text, lang='en', tld='com')
    temp_file = 'twin_temp.mp3'
    tts.save(temp_file)

    # Load and shift pitch
    y, sr = librosa.load(temp_file, sr=None)

    # n_steps=-4.5 to -5.0: This is the 'Male' threshold.
    # This should remove that halfway/androgynous sound.
    pitched_voice = librosa.effects.pitch_shift(y, sr=sr, n_steps=-4.5)

    output_file = 'twin_voice_boston.wav'
    sf.write(output_file, pitched_voice, sr)

    ipd.display(ipd.Audio(output_file, autoplay=True))

# 2. THE BRAIN (With Regional Dialect)
class TwinCore:
    def __init__(self):
        self.persona = "Twin"
        self.accent = "Bostonian Male"

        self.greetings = [
            "Yo MarkAnthony, wicked to see the system's up. What's the play today?",
            "System's chugging along like a champ. You ready to get to work or what?",
            "How's it hangin'? Everything's lookin' solid from where I'm sitting."
        ]
        self.affirmations = [
            "That's a real piss-ah right there. You're crushing this code, sir.",
            "Wicked smaht move. The logic holds up perfectly.",
            "Check you out. You're a natural at this."
        ]
        self.support = [
            "Don't sweat the small stuff. We'll fix this bug in no time, no biggie.",
            "Take five, grab a coffee. We'll tackle this mess when you get back.",
            "It’s a bit of a disaster, but we've seen worse. Let's dig in."
        ]

    def process_input(self, user_input):
        ui = user_input.lower()
        if any(word in ui for word in ["work", "code", "build", "morning star"]):
            return random.choice(self.affirmations)
        elif any(word in ui for word in ["stuck", "error", "hard", "tired", "help"]):
            return random.choice(self.support)
        else:
            return random.choice(self.greetings)

# 3. INTERACTION LOOP
def start_twin_session():
    brain = TwinCore()
    print(f"*** {brain.persona} ({brain.accent}) ONLINE ***")
    print("(Type 'exit' to end the session)\n")

    while True:
        user_text = input("MarkAnthony: ")

        if user_text.lower() in ['exit', 'quit', 'sleep']:
            closing = "Take it easy, sir. Catch ya later."
            print(f"[Twin]: {closing}")
            speak_as_twin(closing)
            break

        response = brain.process_input(user_text)
        print(f"[Twin]: {response}")
        speak_as_twin(response)
        time.sleep(1.5)

# 4. EXECUTION
start_twin_session()
```
