```python
# --- PROJECT (F): NATALIE CONVERSATIONAL CORE (INTEGRATED) ---
# Location: /02_PYTHON_RPI_CODE/RPI_MAIN_PROGRAM/natalie_core.py

import random
import librosa
import soundfile as sf
from gtts import gTTS
import IPython.display as ipd
import time

# Install gTTS if not already installed
!pip install gtts

# 1. NATALIE VOICE ENGINE
def speak_as_natalie(text):
    # Natalie uses the same UK base for her accent
    tts = gTTS(text=text, lang='en', tld='co.uk')
    temp_file = 'natalie_temp.mp3'
    tts.save(temp_file)

    y, sr = librosa.load(temp_file, sr=None)

    # PITCH CALIBRATION:
    # Natalie is 5'6", so we use 0.0 or 0.5 for a natural adult tone.
    # (Unlike BTTRfly's 5.5)
    pitched_voice = librosa.effects.pitch_shift(y, sr=sr, n_steps=0.5)

    output_file = 'natalie_voice.wav'
    sf.write(output_file, pitched_voice, sr)
    ipd.display(ipd.Audio(output_file, autoplay=True))

# 2. NATALIE PERSONA & BRAIN
class NatalieCore:
    def __init__(self):
        self.persona = "NATALie (Project F)"
        self.accent = "Soft Scottish/Irish"

        # Natalie's Archive-accurate responses
        self.greetings = [
            "Ach, hello there, MarkAnthony. It's grand to hear your voice.",
            "Right then, how can I anchor your day?"
        ]
        self.affirmations = [
            "That's brilliant focus, keep that pace. You've got the resolve!",
            "Aye, I see ye've got the determination of a mountain climber."
        ]
        self.support = [
            "Dinnae fash yourself, we'll get through this, step by step.",
            "Take a breath. What's the next small thing we can tackle?"
        ]

    def process_input(self, user_input):
        user_input_lower = user_input.lower()
        if any(word in user_input_lower for word in ["work", "code", "build"]):
            return random.choice(self.affirmations)
        elif any(word in user_input_lower for word in ["stuck", "hard", "tired"]):
            return random.choice(self.support)
        else:
            return random.choice(self.greetings)

# 3. INTERACTION LOOP
def start_natalie_session():
    brain = NatalieCore()
    print(f"*** {brain.persona} ONLINE ***")

    while True:
        user_text = input("MarkAnthony: ")
        if user_text.lower() in ['exit', 'quit', 'sleep']:
            break

        response = brain.process_input(user_text)
        print(f"[{brain.persona}]: {response}")
        speak_as_natalie(response)
        time.sleep(2)

# start_natalie_session()
```
