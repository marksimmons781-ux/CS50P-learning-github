```python
# --- PROJECT (F): NATALIE CONVERSATIONAL CORE (RESTORED V3) ---
# Location: /02_PYTHON_RPI_CODE/RPI_MAIN_PROGRAM/natalie_core.py

!pip install gtts librosa soundfile

import random
import librosa
import soundfile as sf
from gtts import gTTS
import IPython.display as ipd
import time

# 1. NATALIE PERSONA (Restored from Archive V3)
class NataliePersona:
    def __init__(self):
        self.emotion = "calm"
        self.persona_accent = "soft Scottish/Irish"
        self.last_user_statement = ""
        # response_weights logic from V3 is implied here

# 2. THE BRAIN (Integrated with Audio)
class NatalieCore:
    def __init__(self, persona):
        self.state = persona
        # Archive-accurate responses
        self.greetings = [
            "Ach, hello there, MarkAnthony. That's grand to hear your voice.",
            "Right then, how can I anchor your day?"
        ]
        self.affirmations = [
            "That's brilliant focus, keep that pace. You've got the resolve!",
            "Aye, I see ye've got the determination of a mountain climber."
        ]
        self.support = [
            "Dinnae fash yourself, we'll get through this, step by step.",
            "Take a breath. What's the next small thing we can tackle?"
        ]

    def _choose_category(self, user_input):
        """Keyword routing from Archive V3 """
        ui = user_input.lower()
        if any(word in ui for word in ["study", "work", "code", "build"]):
            return "affirmation"
        elif any(word in ui for word in ["problem", "stuck", "hard", "tired"]):
            return "support"
        return "greeting"

    def run_interaction(self, user_input):
        """Generates response and updates state """
        category = self._choose_category(user_input)

        # Select response based on category
        if category == "affirmation":
            response = random.choice(self.affirmations)
        elif category == "support":
            response = random.choice(self.support)
        else:
            response = random.choice(self.greetings)

        self.state.last_user_statement = user_input
        return response

# 3. VOICE ENGINE (Calibrated for 5'6" Adult)
def speak_as_natalie(text):
    tts = gTTS(text=text, lang='en', tld='co.uk')
    temp_file = 'natalie_temp.mp3'
    tts.save(temp_file)

    y, sr = librosa.load(temp_file, sr=None)
    # Natural adult pitch (Archive Standard)
    pitched_voice = librosa.effects.pitch_shift(y, sr=sr, n_steps=0.5)

    output_file = 'natalie_voice.wav'
    sf.write(output_file, pitched_voice, sr)
    ipd.display(ipd.Audio(output_file, autoplay=True))

# 4. MAIN INTERACTION LOOP
def start_natalie_session():
    brain = NatalieCore(NataliePersona())
    print(f"*** NATALie (Project F) ONLINE ***")
    print("(Type 'exit' or 'sleep' to end session)\n")

    while True:
        user_text = input("MarkAnthony: ")

        if user_text.lower() in ['exit', 'quit', 'sleep']:
            closing = "Always remember your anchor, MarkAnthony. Sleep well."
            print(f"[NATALie]: {closing}")
            speak_as_natalie(closing)
            break

        response = brain.run_interaction(user_text)
        print(f"[NATALie]: {response}")
        speak_as_natalie(response)
        time.sleep(2)

# Start Session
start_natalie_session()
```
