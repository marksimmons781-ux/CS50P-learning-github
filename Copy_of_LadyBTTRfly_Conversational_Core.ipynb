```python
# --- LADY BTTRFLY: LIVE INTERACTION TERMINAL ---
# Purpose: Real-time chat with Pixie-Class Pitch Shifting

import random
import librosa
import soundfile as sf
from gtts import gTTS
import IPython.display as ipd
import time

# 1. VOICE ENGINE (The Pitch-Shifter)
def speak_as_bttrfly(text):
    """Generates UK English speech and shifts pitch to a child-like frequency."""
    tts = gTTS(text=text, lang='en', tld='co.uk')
    temp_file = 'temp_voice.mp3'
    tts.save(temp_file)

    # Load and shift pitch (n_steps=5.5 for youthful/pixie tone)
    y, sr = librosa.load(temp_file, sr=None)
    pitched_voice = librosa.effects.pitch_shift(y, sr=sr, n_steps=5.5)

    output_file = 'bttrfly_pixie_voice.wav'
    sf.write(output_file, pitched_voice, sr)

    # Display the audio player (autoplay ensures she speaks immediately)
    ipd.display(ipd.Audio(output_file, autoplay=True))

# 2. THE BRAIN (BTTRfly Core Logic)
class BTTRflyCore:
    def __init__(self):
        self.persona_name = "Lady BTTRfly"
        self.accent = "Pixie-Class Analytical"
        self.greetings = [
            "System online. Analysis ready, MarkAnthony.",
            "I'm seeing the world clearly today. Shall we begin our walk?"
        ]
        self.affirmations = [
            "Data suggests your focus is optimal. Excellent progress.",
            "I've cross-referenced the motor logs. Your determination is quantifiable!"
        ]
        self.support = [
            "Recalibrating approach. We can solve this logically, step by step.",
            "Don't let the noise interfere with the data. What is our next objective?"
        ]

    def process_input(self, user_input):
        user_input_lower = user_input.lower()

        # Keyword routing
        if any(word in user_input_lower for word in ["study", "work", "code", "natalie", "project"]):
            response = random.choice(self.affirmations)
        elif any(word in user_input_lower for word in ["stuck", "error", "problem", "fail", "hard"]):
            response = random.choice(self.support)
        else:
            response = random.choice(self.greetings)

        return response

# 3. MAIN INTERACTION LOOP
def start_chat():
    brain = BTTRflyCore()
    print(f"*** {brain.persona_name} SESSION STARTING ***")
    print("(Type 'exit', 'quit', or 'sleep' to end the session)\n")

    while True:
        # Get input from user
        user_text = input("MarkAnthony: ")

        # Check for exit commands
        if user_text.lower() in ['exit', 'quit', 'sleep', 'shutdown']:
            closing = "Understood. Powering down to standby mode. Sleep well, MarkAnthony."
            print(f"[{brain.persona_name}]: {closing}")
            speak_as_bttrfly(closing)
            break

        # Process and respond
        response = brain.process_input(user_text)
        print(f"[{brain.persona_name}]: {response}")

        # Trigger audio
        speak_as_bttrfly(response)

        # Brief pause to prevent overlapping audio issues in Colab
        time.sleep(2)

# Start the program
start_chat()
```
