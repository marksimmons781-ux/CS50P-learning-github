```python
import pandas as pd

#Define the structure for my story collection
stories_data = {
    'story_id':['DragonLands','nebula_echo','Deception','Developing_Attractions','Constructive_Critisim','Autumn_Rayne', 'Vampires_Alive', 'Unidentified', 'Fog', 'Cry_For_Help', 'Where_Are_You_Now', 'Keri_Come_Back', 'The_New_Kid', '6E', 'World_On_Fire'],
    'title':['DragonLands','The Nebula Echo', 'Deception', 'Developing Attractions', 'Constructive Critisim', 'Autumn Rayne', 'Vampires Alive', 'Unidentified', 'Fog', 'Cry For Help', 'Where Are You Now', 'Keri Come Back', 'The New Kid', '6E', 'World On Fire'],
    'genre':['High Fantasy','Sci-Fi', 'Psychological Thriller', 'Romance', 'Romance', 'Romance', 'Sci-fi Horror', 'Murder Mystery', 'Psycological Horror', 'YA', 'YA', 'YA', 'YA', 'Murder Mystery', 'Social Critique Non-Fiction'], # Now 15 elements
    'plot_outline':['A world where dragons correct the earth, focusing on the survival of Great Harbor and Volcanis.',
                    'A Lone signal from a dead star leads a crew into a temporal rift.',
                    'A Woman Detective is investigating a string of murders which appear to be connected with a similar one from her past.',
                    'Seeing her boyfriend in a diner sitting at the spot that she claimed was "theirs" Gwen makes the the mistake of assuming he was cheating and causes a devistating outcome in their relationship.',
                    'Gwen must fight to find and reclaim the love of her life but is it already too late or can she reconstruct the love once was again?',
                    'Betrayed by her caregivers and kidnapped by a rugged pirate, Autumn struggles to find her place and a new home free from the control of others.',
                    '', '', '', '', '', '', '', '', ''] # Now 15 elements
}

#Create the DataFrame
df_Stories = pd.DataFrame(stories_data)

#Display the current library of stories
print("---Story Master List---")
display(df_Stories)

# 1. Define the structure for Characters
# We include 'profile' and 'backstory' as separate columns per your request
chars_data = {
    'char_id': ['Rori_ID','Tom-Tom_ID','Collin_Thornwood_ID'],
    'story_id': ['DragonLands', 'DragonLands', 'DragonLands'], # Must match the IDs in your first table (e.g., 'DragonLands')
    'name': ['Rori','Tom-Tom','Collin Thornwood'],
    'role': ['Protagonist','Support','Support'],
    'profile': ['Orphaned girl','Guardian','Lord of Great Harbor'],
    'backstory': ['', '', '']
}

df_Characters = pd.DataFrame(chars_data)

# 2. A function to make adding characters faster
def add_char(c_id, s_id, name, role, profile, backstory):
    global df_Characters
    new_entry = {
        'char_id': c_id,
        'story_id': s_id,
        'name': name,
        'role': role,
        'profile': profile,
        'backstory': backstory
    }
    df_Characters = pd.concat([df_Characters, pd.DataFrame([new_entry])], ignore_index=True)

# 3. Adding your initial characters
add_char('Rori_01', 'DragonLands', 'Rori', 'Protagonist', 'Young, resilient', 'Lost parents to dragon fire.')
add_char('Autumn_01', 'Autumn_Rayne', 'Autumn Rayne', 'Protagonist', 'Determined, captive', 'Betrayed by caregivers.')
add_char('Pirate_01', 'Autumn_Rayne', 'Justin Black', 'Antagonist/Love Interest', 'Rugged, mysterious', 'Kidnapped Autumn.')
add_char('Kara_West_ID', 'Deception', 'Kara West', 'Protagonist', 'Detective', 'Investigating a string of murders which have a striking resemblance to one from her own past')
add_char('Dillon_Carter_ID', 'Deception', 'Dillon Carter', 'Support', 'Detective', 'Transferred in from big city must prove to Kara that he is on her side')
add_char('Edward_Tomlin_ID', "Vampires_Alive", 'Edward Tomlin', 'Protagonist', 'Survivor', 'Excaped from Vampire feeding colonies struggling to remain alive and free')
add_char('Garret_Holstein_ID', 'DragonLands', 'Garret Holstein', 'Antagonist', 'Rapist/murderer/soldier/outlaw', 'Garret is the leader of the former soldier group turned outlaws in the Midland area')
add_char('Lorrent_ID','DragonLands', 'Lorrent', 'Antagonist','Outlaw/murderer/rapist/missionary','Lorrent is the second in command of Garrets team he will later become resentful of his past and seek ammends')
add_char('Tristane_ID','DragonLands','Tristane','Protector', 'Former Squire','Squire turned knight and sworn sword to Collin Thornwood')

print("--- Character Profiles Updated ---")
display(df_Characters)

# Identify stories with missing genre or plot_outline
missing_info_stories = df_Stories[(df_Stories['genre'] == '') | (df_Stories['plot_outline'] == '')]

# Display the stories with missing information
print("---Stories with Missing Genre or Plot Outline---")
display(missing_info_stories)

# Re-create df_Master_View by merging df_Characters and df_Stories
df_Master_View = pd.merge(df_Characters, df_Stories, on='story_id', how='left')

# Define romance_stories DataFrame
romance_stories = df_Stories[df_Stories['genre'] == 'Romance']

# Add 'is_romance' column to df_Master_View
# This checks if the story_id in df_Master_View is present in the romance_stories DataFrame
df_Master_View['is_romance'] = df_Master_View['story_id'].isin(romance_stories['story_id'])

print("---Unified df_Master_View---")
display(df_Master_View.head())


# Create a dictionary to map story titles to their trigger warnings
trigger_warning_map = {
    'DragonLands': ['Visceral Violence', 'Existential Dread'],
    'The Monster Within': ['Abuse', 'Domestic Trauma'], # Assuming 'The Monster Within' exists in your stories
    'Fairmont': ['None'], # Assuming 'Fairmont' exists in your stories
    'Fog': ['Gore', 'Psychological Horror']
}

# Add the 'Trigger_Warnings' column to df_Master_View by mapping titles
# For titles not in the map, assign ['None'] as a default
df_Master_View['Trigger_Warnings'] = df_Master_View['title'].map(trigger_warning_map).apply(lambda x: x if isinstance(x, list) else ['None'])
```

## Analyze Genre and Role Distribution

### Subtask:
Calculate and display the distribution of genres and character roles within the `df_Master_View`.


**Reasoning**:
To analyze the distribution of genres and character roles, I will use the `value_counts()` method on the 'genre' and 'role' columns of `df_Master_View` and then display the results with appropriate headers.



```python
print("---Genre Distribution---")
genre_distribution = df_Master_View['genre'].value_counts()
display(genre_distribution)

print("\n---Character Role Distribution---")
role_distribution = df_Master_View['role'].value_counts()
display(role_distribution)
```

## Analyze Trigger Warning Distribution

### Subtask:
Examine the frequency of different trigger warnings assigned to stories in `df_Master_View`.


**Reasoning**:
To analyze the distribution of trigger warnings, I will iterate through the 'Trigger_Warnings' column, count the occurrences of each unique warning, and then display these counts as a sorted Pandas Series.



```python
import collections

print("---Trigger Warning Distribution---")

# Initialize an empty dictionary to store the counts of each unique trigger warning
trigger_warning_counts = collections.defaultdict(int)

# Iterate through each list of trigger warnings in the 'Trigger_Warnings' column
for warnings_list in df_Master_View['Trigger_Warnings']:
    # For each warning in each list, update its count
    for warning in warnings_list:
        trigger_warning_counts[warning] += 1

# Convert the trigger_warning_counts dictionary into a Pandas Series and sort it
warning_distribution = pd.Series(trigger_warning_counts).sort_values(ascending=False)

# Display the resulting warning_distribution Series
display(warning_distribution)
```

## Visualize Key Distributions

### Subtask:
Generate bar charts to visualize the distribution of genres, character roles, and trigger warnings. Ensure all plots have appropriate titles and legends.


**Reasoning**:
I will import the necessary libraries for plotting, `matplotlib.pyplot` and `seaborn`, to prepare for creating the visualizations.



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Reasoning**:
Now that the plotting libraries are imported, I will generate bar charts for the distribution of genres, character roles, and trigger warnings, ensuring each plot has an appropriate title and axis labels as specified in the instructions.



```python
plt.figure(figsize=(10, 6))
sns.barplot(x=genre_distribution.index, y=genre_distribution.values)
plt.title('Distribution of Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Stories')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=role_distribution.index, y=role_distribution.values)
plt.title('Distribution of Character Roles')
plt.xlabel('Character Role')
plt.ylabel('Number of Characters')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=warning_distribution.index, y=warning_distribution.values)
plt.title('Distribution of Trigger Warnings')
plt.xlabel('Trigger Warning Type')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
