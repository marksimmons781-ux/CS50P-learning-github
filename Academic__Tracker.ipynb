### Data Cleaning
Next, I'll clean the `Progress` and `Score` columns by converting their percentage strings to numerical floats, and the `Date_Completed` column to datetime objects for easier analysis.

```python
# 2. Data Cleaning
# Convert percentage strings (e.g., '100.00%') to floats (1.0)
def clean_percentage(val):
    if isinstance(val, str):
        try:
            return float(val.strip('%')) / 100
        except ValueError:
            return None # Handle invalid percentage strings
    return val

# Apply the cleaning function to 'Progress' and 'Score' columns
df['Progress_Value'] = df['Progress'].apply(clean_percentage)
df['Score_Value'] = df['Score'].apply(clean_percentage)

# Convert Date_Completed to datetime objects for time-based analysis
df['Date_Completed'] = pd.to_datetime(df['Date_Completed'], errors='coerce')

print("Data Cleaning complete. Displaying info on cleaned columns and the DataFrame head:")
display(df[['Progress', 'Progress_Value', 'Score', 'Score_Value', 'Date_Completed']].head())
display(df.info())
```

### Distribution of Imputed CPD Credit Hours

Now that we've imputed the missing values in `CPD Credit Hrs.`, let's visualize the distribution of the new `CPD_Credit_Hrs_Imputed` column. This will give us a complete picture of credit hour distribution, including the impact of the imputation.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['CPD_Credit_Hrs_Imputed'], bins=8, kde=True, color='green')
plt.title('Distribution of Imputed CPD Credit Hours')
plt.xlabel('CPD Credit Hours (Imputed)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Calculate the median of 'CPD Credit Hrs.'
median_cpd_credit_hrs = df['CPD Credit Hrs.'].median()

# Impute missing values using the median and create a new column
df['CPD_Credit_Hrs_Imputed'] = df['CPD Credit Hrs.'].fillna(median_cpd_credit_hrs)

# Verify that missing values have been imputed
print(f"Number of missing 'CPD Credit Hrs.' values BEFORE imputation: {df['CPD Credit Hrs.'].isnull().sum()}")
print(f"Number of missing 'CPD_Credit_Hrs_Imputed' values AFTER imputation: {df['CPD_Credit_Hrs_Imputed'].isnull().sum()}\n")

# Display the head of the DataFrame to show the new column
print("First 5 rows with the new imputed column:")
display(df[['CPD Credit Hrs.', 'CPD_Credit_Hrs_Imputed']].head())
```

### Relationship between Score and Cognitive Load

Now, let's explore the relationship between the `Score_Value` (the score achieved in a course) and `Cognitive_Load` (how cognitively demanding the course was). A scatter plot is suitable for this to observe any potential correlations.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cognitive_Load', y='Score_Value', data=df, hue='Score_Value', size='Score_Value', sizes=(20, 400), palette='viridis', legend='full')
plt.title('Score vs. Cognitive Load')
plt.xlabel('Cognitive Load')
plt.ylabel('Score (Value)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
